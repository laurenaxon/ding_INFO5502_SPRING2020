```
import numpy as np
import matplotlib.pyplot as plt



#Code Outline for All Subsequent Figures


#Plotting Random Points
N = 10
"""Changing N changes the number of points created"""

x = numpy.random.uniform(low=-1, high=1, size=[N,1]) 
y = numpy.random.uniform(low=-1, high=1, size=[N,1])


#Turning the Inside Points Red
inside_bool=x**2 + y**2<1

x_in=0.5*x[inside_bool] + 0.5
y_in=0.5*y[inside_bool] + 0.5

x= 0.5*x+0.5
y= 0.5*y+0.5

"""The *0.5 scaled the circle down to have a radius of 0.5"""
"""The +0.5 shifted the x & y range so the circle centered over (0.5,0.5)"""
#Creating 1x1 Plot
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)

#Creating Circle Inside Plot
ax = plt.axes()

ax.set_aspect(1)
theta = np.linspace(-np.pi, np.pi, 100)
plt.plot((0.5*np.sin(theta) + 0.5), (0.5*np.cos(theta) + 0.5), color='silver')

plt.scatter(x,y,s=2)
plt.scatter(x_in,y_in, color='red', s=2)
plt.title('N=10')
plt.show()

approximation10=4*np.sum(inside_bool)/N
print('Pi: {}, Approximation: {}'.format(np.pi, approximation10))






#Rinse & Repeat
"""Copy & Pasted the above code and consolidated by removing spacer lines and commentary. 
Then changed N= ? and the name on the approximation values to reflect the different N level"""




#N=500

N = 500

x = numpy.random.uniform(low=-1, high=1, size=[N,1]) 
y = numpy.random.uniform(low=-1, high=1, size=[N,1])
inside_bool=x**2 + y**2<1
approximation=4*np.sum(inside_bool)/N 
x_in=0.5*x[inside_bool] + 0.5
y_in=0.5*y[inside_bool] + 0.5
x= 0.5*x+0.5
y= 0.5*y+0.5
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)
ax = plt.axes()
ax.set_aspect(1)
theta = np.linspace(-np.pi, np.pi, 100)
plt.plot((0.5*np.sin(theta) + 0.5), (0.5*np.cos(theta) + 0.5), color='silver')
plt.scatter(x,y,s=2)
plt.scatter(x_in,y_in, color='red', s=2)
plt.title('N=500')
plt.show()

approximation500=4*np.sum(inside_bool)/N
print('Pi: {}, Approximation: {}'.format(np.pi, approximation500))



#N=1,000

N = 1000

x = numpy.random.uniform(low=-1, high=1, size=[N,1]) 
y = numpy.random.uniform(low=-1, high=1, size=[N,1])
inside_bool=x**2 + y**2<1
approximation=4*np.sum(inside_bool)/N 
x_in=0.5*x[inside_bool] + 0.5
y_in=0.5*y[inside_bool] + 0.5
x= 0.5*x+0.5
y= 0.5*y+0.5
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)
ax = plt.axes()
ax.set_aspect(1)
theta = np.linspace(-np.pi, np.pi, 100)
plt.plot((0.5*np.sin(theta) + 0.5), (0.5*np.cos(theta) + 0.5), color='silver')
plt.scatter(x,y,s=2)
plt.scatter(x_in,y_in, color='red', s=2)
plt.title('N=1,000')
plt.show()

approximation1k=4*np.sum(inside_bool)/N
print('Pi: {}, Approximation: {}'.format(np.pi, approximation1k))



#N=5,000

N = 5000

x = numpy.random.uniform(low=-1, high=1, size=[N,1]) 
y = numpy.random.uniform(low=-1, high=1, size=[N,1])
inside_bool=x**2 + y**2<1
approximation=4*np.sum(inside_bool)/N 
x_in=0.5*x[inside_bool] + 0.5
y_in=0.5*y[inside_bool] + 0.5
x= 0.5*x+0.5
y= 0.5*y+0.5
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)
ax = plt.axes()
ax.set_aspect(1)
theta = np.linspace(-np.pi, np.pi, 100)
plt.plot((0.5*np.sin(theta) + 0.5), (0.5*np.cos(theta) + 0.5), color='silver')
plt.scatter(x,y,s=2)
plt.scatter(x_in,y_in, color='red', s=2)
plt.title('N=5,000')
plt.show()

approximation5k=4*np.sum(inside_bool)/N
print('Pi: {}, Approximation: {}'.format(np.pi, approximation5k))



#N=10,000

N = 10000

x = numpy.random.uniform(low=-1, high=1, size=[N,1]) 
y = numpy.random.uniform(low=-1, high=1, size=[N,1])
inside_bool=x**2 + y**2<1
approximation=4*np.sum(inside_bool)/N 
x_in=0.5*x[inside_bool] + 0.5
y_in=0.5*y[inside_bool] + 0.5
x= 0.5*x+0.5
y= 0.5*y+0.5
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)
ax = plt.axes()
ax.set_aspect(1)
theta = np.linspace(-np.pi, np.pi, 100)
plt.plot((0.5*np.sin(theta) + 0.5), (0.5*np.cos(theta) + 0.5), color='silver')
plt.scatter(x,y,s=2)
plt.scatter(x_in,y_in, color='red', s=2)
plt.title('N=10,000')
plt.show()

approximation10k=4*np.sum(inside_bool)/N
print('Pi: {}, Approximation: {}'.format(np.pi, approximation10k))



#N=50,000

N = 50000

x = numpy.random.uniform(low=-1, high=1, size=[N,1]) 
y = numpy.random.uniform(low=-1, high=1, size=[N,1])
inside_bool=x**2 + y**2<1
approximation=4*np.sum(inside_bool)/N 
x_in=0.5*x[inside_bool] + 0.5
y_in=0.5*y[inside_bool] + 0.5
x= 0.5*x+0.5
y= 0.5*y+0.5
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)
ax = plt.axes()
ax.set_aspect(1)
theta = np.linspace(-np.pi, np.pi, 100)
plt.plot((0.5*np.sin(theta) + 0.5), (0.5*np.cos(theta) + 0.5), color='silver')
plt.scatter(x,y,s=2)
plt.scatter(x_in,y_in, color='red', s=2)
plt.title('N=50,000')
plt.show()

approximation50k=4*np.sum(inside_bool)/N
print('Pi: {}, Approximation: {}'.format(np.pi, approximation50k))



#N=80,000

N = 80000

x = numpy.random.uniform(low=-1, high=1, size=[N,1]) 
y = numpy.random.uniform(low=-1, high=1, size=[N,1])
inside_bool=x**2 + y**2<1
approximation=4*np.sum(inside_bool)/N 
x_in=0.5*x[inside_bool] + 0.5
y_in=0.5*y[inside_bool] + 0.5
x= 0.5*x+0.5
y= 0.5*y+0.5
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)
ax = plt.axes()
ax.set_aspect(1)
theta = np.linspace(-np.pi, np.pi, 100)
plt.plot((0.5*np.sin(theta) + 0.5), (0.5*np.cos(theta) + 0.5), color='silver')
plt.scatter(x,y,s=2)
plt.scatter(x_in,y_in, color='red', s=2)
plt.title('N=80,000')
plt.show()

approximation80k=4*np.sum(inside_bool)/N
print('Pi: {}, Approximation: {}'.format(np.pi, approximation80k))



#N=100,000

N = 100000

x = numpy.random.uniform(low=-1, high=1, size=[N,1]) 
y = numpy.random.uniform(low=-1, high=1, size=[N,1])
inside_bool=x**2 + y**2<1
approximation=4*np.sum(inside_bool)/N 
x_in=0.5*x[inside_bool] + 0.5
y_in=0.5*y[inside_bool] + 0.5
x= 0.5*x+0.5
y= 0.5*y+0.5
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)
ax = plt.axes()
ax.set_aspect(1)
theta = np.linspace(-np.pi, np.pi, 100)
plt.plot((0.5*np.sin(theta) + 0.5), (0.5*np.cos(theta) + 0.5), color='silver')
plt.scatter(x,y,s=2)
plt.scatter(x_in,y_in, color='red', s=2)
plt.title('N=100,000')
plt.show()

approximation100k=4*np.sum(inside_bool)/N
print('Pi: {}, Approximation: {}'.format(np.pi, approximation100k))



#N=200,000

N = 200000

x = numpy.random.uniform(low=-1, high=1, size=[N,1]) 
y = numpy.random.uniform(low=-1, high=1, size=[N,1])
inside_bool=x**2 + y**2<1
approximation=4*np.sum(inside_bool)/N 
x_in=0.5*x[inside_bool] + 0.5
y_in=0.5*y[inside_bool] + 0.5
x= 0.5*x+0.5
y= 0.5*y+0.5
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)
ax = plt.axes()
ax.set_aspect(1)
theta = np.linspace(-np.pi, np.pi, 100)
plt.plot((0.5*np.sin(theta) + 0.5), (0.5*np.cos(theta) + 0.5), color='silver')
plt.scatter(x,y,s=2)
plt.scatter(x_in,y_in, color='red', s=2)
plt.title('N=200,000')
plt.show()

approximation200k=4*np.sum(inside_bool)/N
print('Pi: {}, Approximation: {}'.format(np.pi, approximation200k))






#Line chart
approx = [approximation10,approximation500,approximation1k,approximation5k,approximation10k,approximation50k,approximation80k,approximation100k,approximation200k]
approx_array = np.array(a)

error = abs(np.pi - ar)
error_array = np.array(b) 

n=[10,500,1000,5000,10000,50000,80000,100000,200000]

plt.plot(n, error_array)
plt.title('Approximation Error by Number of Points')
plt.xlabel('The Total Number of Plot Points (N=?)')
plt.ylabel('The Abs. Value of Pi - The Approximation')
plt.show()
```
