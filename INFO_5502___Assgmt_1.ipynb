```
import random
import string

chars = string.ascii_uppercase + string.digits
from random import randrange
A = [''.join(random.choices(string.ascii_uppercase + string.digits, k=randrange(1,11))) for j in range(500)]

'''A = [...] means there was a new list created titled "A"
  random.choices means that given the following parameters, each string sill be created at random
    1.) The strings will have uppercase letters and numbers
    2.) The strings will be between 1 and 10 characters long
  j in range(500) means the list will have 500 strings'''


print(A)

def bubbleSort(A):
    for passnum in range(len(A)-1,0,-1):
        for i in range(passnum):
            if A[i]>A[i+1]:
                temp = A[i]
                A[i] = A[i+1]
                A[i+1] = temp

bubbleSort(A)

"""The Sorting method used here is called bubble sorting in which the second string in this list is compared to the first string in this list.
If the second string comes before the first string alphanumerically, then the two are swapped making the second the string the first and vice versa.
Then the third string is compared to the now second and the process repeats.
This process loops through the list until there are no swaps to be made. 

In this case, digits are sorted in front of letters."""


print(A)


#Output line 1: Original List
#Output line 2: Sorted List


CountA = tuple(len(a) for a in A)
print(CountA)

#Output line 3: The number of characters for each string in Sorted List A

import matplotlib.pyplot as plt
import numpy as np

y= [list(range(1,501))]
x= [CountA]

plt.scatter(x,y, label='scatterplot', color='k', s=25, marker="o")

plt.xlabel('String Length')
plt.ylabel('String Position in List')
plt.title('Assignment 1\nString Length')
plt.legend()
plt.show()
```
